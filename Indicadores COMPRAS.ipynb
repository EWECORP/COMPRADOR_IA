## Modificaciones agregadas:
1) Líneas de referencia (axhline o axvline) en los gráficos donde es relevante comparar valores contra un objetivo.
2) Leyendas (legend()) para indicar la meta o referencia en cada gráfico.
3) Diferentes valores de referencia:
* Cobertura de Stock: 25 días.
* Órdenes Pendientes: Máximo 8 órdenes.
* Cumplimiento OTIF: 85%.
* Alertas de Stock Crítico: Máximo 2 alertas.
* Evolución de Deuda: Referencia en 1000$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear la figura para el panel de gestión
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Panel de Gestión de Compradores", fontsize=16, weight='bold')
fig.tight_layout(pad=5.0)

# Datos para los gráficos
categories = ['Cat1', 'Cat2', 'Cat3', 'Cat4']
values = [20, 15, 35, 30]

# Línea de referencia (valor objetivo)
stock_target = 25
pending_orders_target = 8
otif_target = 85
critical_stock_target = 2
debt_target = 1000

# Gráfico 1: Cobertura de Stock
axs[0, 0].bar(categories, values, color='skyblue')
axs[0, 0].axhline(stock_target, color='red', linestyle='--', linewidth=2, label=f"Meta ({stock_target} días)")
axs[0, 0].set_title("Cobertura de Stock (días)")
axs[0, 0].set_ylabel("Días")
axs[0, 0].set_xlabel("Categorías")
axs[0, 0].legend()

# Gráfico 2: Órdenes Pendientes de Aprobación
pending_orders = [5, 10, 15, 7]
axs[0, 1].bar(categories, pending_orders, color='lightcoral')
axs[0, 1].axhline(pending_orders_target, color='red', linestyle='--', linewidth=2, label=f"Máximo ({pending_orders_target} órdenes)")
axs[0, 1].set_title("Órdenes Pendientes de Aprobación")
axs[0, 1].set_ylabel("Número de Órdenes")
axs[0, 1].set_xlabel("Proveedores")
axs[0, 1].legend()

# Gráfico 3: Cumplimiento OTIF
otif = [90, 80, 70, 95]
axs[0, 2].bar(categories, otif, color='lightgreen')
axs[0, 2].axhline(otif_target, color='red', linestyle='--', linewidth=2, label=f"Objetivo ({otif_target}%)")
axs[0, 2].set_title("Cumplimiento OTIF (%)")
axs[0, 2].set_ylabel("%")
axs[0, 2].set_xlabel("Proveedores")
axs[0, 2].legend()

# Gráfico 4: Utilización del Presupuesto (No requiere línea de referencia)
axs[1, 0].pie([70, 30], labels=['Usado', 'Disponible'], autopct='%1.1f%%', colors=['gold', 'lightgray'])
axs[1, 0].set_title("Utilización del Presupuesto")

# Gráfico 5: Alertas de Stock Crítico
critical_stock = [3, 4, 2, 1]
axs[1, 1].barh(categories, critical_stock, color='salmon')
axs[1, 1].axvline(critical_stock_target, color='red', linestyle='--', linewidth=2, label=f"Máximo ({critical_stock_target} alertas)")
axs[1, 1].set_title("Alertas de Stock Crítico")
axs[1, 1].set_xlabel("Número de Productos")
axs[1, 1].legend()

# Gráfico 6: Evolución de Deuda con Proveedores
time = np.arange(1, 6)
debt_values = [1000, 1200, 1100, 900, 950]
axs[1, 2].plot(time, debt_values, marker='o', color='blue')
axs[1, 2].axhline(debt_target, color='red', linestyle='--', linewidth=2, label=f"Referencia ({debt_target}$)")
axs[1, 2].set_title("Evolución de Deuda con Proveedores")
axs[1, 2].set_ylabel("Deuda ($)")
axs[1, 2].set_xlabel("Meses")
axs[1, 2].legend()

# Mostrar el dashboard
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Sample data for the product grid
data = {
    "Código": ["001", "002", "003", "004"],
    "Producto": ["Coca-Cola 500ml", "Galletitas X", "Pepsi 1L", "Snacks Y"],
    "Categoría": ["Bebidas", "Snacks", "Bebidas", "Snacks"],
    "Stock Actual": [200, 50, 120, 80],
    "Stock Crítico": [50, 30, 60, 40],
    "Sugerido": [100, 20, 50, 30],
    "Precio Unitario": [50, 30, 45, 20],
    "Promoción": ["10% Descuento", "Ninguna", "5% Descuento", "Ninguna"],
    "Costo Total": [5000, 600, 2250, 600],
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a plot to display the table
fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('off')  # Remove plot axes

# Create the table
tbl = table(ax, df, loc='center', cellLoc='center', colWidths=[0.12] * len(df.columns))

# Style the table
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.auto_set_column_width(col=list(range(len(df.columns))))

# Highlight certain rows/columns (optional)
for (row, col), cell in tbl.get_celld().items():
    if row == 0:  # Header
        cell.set_fontsize(12)
        cell.set_text_props(weight='bold')
        cell.set_facecolor("#d9f2d9")  # Light green for headers
    elif col == 4 and row > 0:  # Stock Crítico
        if df.iloc[row - 1, col - 1] > df.iloc[row - 1, 3]:  # Check if stock is below critical
            cell.set_facecolor("#f2d9d9")  # Light red for critical stock

plt.title("Grilla de Productos Sugeridos para Reposición", fontsize=14, weight='bold', pad=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Sample data for the product grid
data = {
    "Código": ["001", "002", "003", "004"],
    "Producto": ["Coca-Cola 500ml", "Galletitas X", "Pepsi 1L", "Snacks Y"],
    "Categoría": ["Bebidas", "Snacks", "Bebidas", "Snacks"],
    "Stock Actual": [200, 50, 120, 80],
    "Stock Crítico": [50, 30, 60, 40],
    "Sugerido": [100, 20, 50, 30],
    "Confirmado": [0, 0, 0, 0],
    "Precio Unitario": [50, 30, 45, 20],
    "Promoción": ["10% Descuento", "Ninguna", "5% Descuento", "Ninguna"],
    "Costo Total Sugerido": [5000, 600, 2250, 600],
    "Costo Total Confirmado": [1000, 0, 0, 0],
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate summary totals
total_sugerido = df["Costo Total Sugerido"].sum()
total_confirmado = df["Costo Total Confirmado"].sum()

# Create the figure and axes
fig, (ax_summary, ax_grid) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [1, 4]})
fig.patch.set_facecolor("#f5f5f5")

# Upper summary panel
ax_summary.set_facecolor("#d9f2d9")
ax_summary.axis("off")
ax_summary.text(0.2, 0.5, f"Total Sugerido: ${total_sugerido:,}", fontsize=12, weight="bold", ha="center")
ax_summary.text(0.8, 0.5, f"Total Confirmado: ${total_confirmado:,}", fontsize=12, weight="bold", ha="center")

# Lower grid panel
ax_grid.axis("off")
tbl = table(ax_grid, df, loc='center', cellLoc='center', colWidths=[0.08] * len(df.columns))

# Style the table
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.auto_set_column_width(col=list(range(len(df.columns))))

# Highlight stock critical
for (row, col), cell in tbl.get_celld().items():
    if row == 0:  # Header
        cell.set_fontsize(12)
        cell.set_text_props(weight="bold")
        cell.set_facecolor("#b3cde0")  # Light blue for headers
    elif col == 4 and row > 0:  # Stock Crítico check
        if df.iloc[row - 1, col - 1] > df.iloc[row - 1, 3]:  # Check if stock is below critical
            cell.set_facecolor("#f2d9d9")  # Light red for critical stock

# Title
plt.suptitle("Grilla de Productos Sugeridos para Reposición", fontsize=14, weight="bold", color="#333333", y=0.95)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table


# Recreate data with additional metrics
data = {
    "Código": ["001", "002", "003", "004"],
    "Producto": ["Coca-Cola 500ml", "Galletitas X", "Pepsi 1L", "Snacks Y"],
    "Categoría": ["Bebidas", "Snacks", "Bebidas", "Snacks"],
    "Stock Actual": [200, 50, 120, 80],
    "Stock Crítico": [50, 30, 60, 40],
    "Sugerido": [100, 20, 50, 30],
    "Confirmado": [0, 0, 0, 0],
    "Rotación Mensual": [300, 100, 200, 150],
    "Lead Time (Días)": [7, 5, 10, 6],
    "Precio Unitario": [50, 30, 45, 20],
    "Promoción": ["10% Descuento", "Ninguna", "5% Descuento", "Ninguna"],
    "Costo Total Sugerido": [5000, 600, 2250, 600],
    "Costo Total Confirmado": [0, 0, 0, 0],
}

# Create updated DataFrame
df = pd.DataFrame(data)

# Calculate summary totals
total_sugerido = df["Costo Total Sugerido"].sum()
total_confirmado = df["Costo Total Confirmado"].sum()
critical_products = sum(df["Stock Actual"] <= df["Stock Crítico"])

# Create the figure and axes
fig, (ax_summary, ax_grid) = plt.subplots(2, 1, figsize=(14, 10), gridspec_kw={'height_ratios': [1, 4]})
fig.patch.set_facecolor("#f5f5f5")

# Upper summary panel
ax_summary.set_facecolor("#d9f2d9")
ax_summary.axis("off")
ax_summary.text(0.2, 0.5, f"Total Sugerido: ${total_sugerido:,}", fontsize=12, weight="bold", ha="center")
ax_summary.text(0.5, 0.5, f"Total Confirmado: ${total_confirmado:,}", fontsize=12, weight="bold", ha="center")
ax_summary.text(0.8, 0.5, f"Productos Críticos: {critical_products}", fontsize=12, weight="bold", ha="center")

# Lower grid panel
ax_grid.axis("off")
tbl = table(ax_grid, df, loc='center', cellLoc='center', colWidths=[0.08] * len(df.columns))

# Style the table
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.auto_set_column_width(col=list(range(len(df.columns))))

# Highlight stock critical
for (row, col), cell in tbl.get_celld().items():
    if row == 0:  # Header
        cell.set_fontsize(12)
        cell.set_text_props(weight="bold")
        cell.set_facecolor("#b3cde0")  # Light blue for headers
    elif col == 4 and row > 0:  # Stock Crítico check
        if df.iloc[row - 1, col - 1] > df.iloc[row - 1, 3]:  # Check if stock is below critical
            cell.set_facecolor("#f2d9d9")  # Light red for critical stock

# Title
plt.suptitle("Grilla de Productos Sugeridos para Reposición", fontsize=14, weight="bold", color="#333333", y=0.95)

plt.tight_layout()
plt.show()
```
